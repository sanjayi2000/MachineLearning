```python
import cv2
import matplotlib.pyplot as plt
```

```python
tasm = cv2.imread('/content/wallpaperflare.com_wallpaper.jpg')
```

```python
median_filtered_image = cv2.medianBlur(tasm, 5)
```

```python
bilateral_filtered_image = cv2.bilateralFilter(tasm, 9, 75, 75)
```

```python
denoised_image = cv2.fastNlMeansDenoisingColored(tasm, None, 10, 10, 7, 21)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(141), plt.imshow(tasm)
plt.title('Input Image'), plt.xticks([]), plt.yticks([])

plt.subplot(142), plt.imshow(median_filtered_image)
plt.title('Median Filtered Image'), plt.xticks([]), plt.yticks([])

plt.subplot(143), plt.imshow(bilateral_filtered_image)
plt.title('Bilateral Filtered Image'), plt.xticks([]), plt.yticks([])

plt.subplot(144), plt.imshow(denoised_image)
plt.title('Denoised Image'), plt.xticks([]), plt.yticks([])

plt.show()
```
